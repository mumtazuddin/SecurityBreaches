## Hypothesis I

- **Do you have any expectation about the number of events reported for each year?**

   **Is the trend upwards or downwards?**

```python
plt.subplots(figsize=(12,10))
sns.countplot(df['year'])
#this plot is a countplot which shows breaches yearwise 
```

By this plot we can easily prove the hypothesis that the trend is upwards from the start which is **1997** uptil **2013** breaches has risen up from just **1 in 1997** to almost **254 in 2013** which is clearly showing an upward trend of the breaches happening over the year in different states in different years and then we can see a significant fall in the year **2014** to almost **56 in 2014**  i guess which is because of taking preventing measures and security measures to bypass the breaches and keeping the data secure.

 This proves Hypothesis I

## Hypothesis II

 
 -**Do you have any expectation about the state in which breaches occurred?**

 -**Which states are the largest and therefore might have the most breaches?**

```python
plt.subplots(figsize=(15,10))
sns.countplot(df['State'],order = df['State'].value_counts().index )
#this shows the count of states in which breaches occured and where the most breaches occured 
```

Well this countplot shows the trend in the different states on breaches over the time which i have arranged in a descending order of the highest to lowest so we can see that the most breaches are in **state of california** which is a fairly bigger state than others hence more breaches in that state and then the trend follows to the second biggest state area wise which is **state of texas** and then you can easily see in the plot that the last in the place is a fairly small **state of vermont** and hence very less breaches there less than 5. 

This proves the Hypothesis- II
